```python
import pandas as pd

# Group by 'name' (team) and 'season', then calculate the cumulative win percentage
df['past_wins'] = (
    df.groupby(['name', 'season'])['win_indicator']
    .cumsum()
    .shift(1)  # Shift by 1 to exclude the current game
    .fillna(0)  # Replace NaN for the first game
)

df['past_games'] = df.groupby(['name', 'season']).cumcount()  # Count past games (excluding current)

# Avoid division by zero
df['running_win_pct'] = df['past_wins'] / df['past_games']
df['running_win_pct'] = df['running_win_pct'].fillna(0).round(3)  # Replace NaN with 0 for first game

# No need to drop duplicates
print(df[['name', 'season', 'running_win_pct']].head(7))  # Keep all rows
```

```python
import pandas as pd

# Ensure DataFrame is sorted by 'name', 'season', and 'Game_Number'
df = df.sort_values(['name', 'season', 'Game_Number']).reset_index(drop=True)

# Calculate the 3-game moving average for numeric columns
numeric_cols = df.select_dtypes(include='number').columns
for col in numeric_cols:
    df[f'{col}_3MA'] = (
        df.groupby(['name', 'season'])[col]
        .transform(lambda x: x.rolling(window=3, min_periods=1).mean())
    )

# Handle non-numeric columns (shift to align with the 3rd game onward)
non_numeric_cols = df.select_dtypes(exclude='number').columns
for col in non_numeric_cols:
    df[f'{col}_3MA'] = (
        df.groupby(['name', 'season'])[col]
        .shift(-2)
    )

# Drop the first two games per 'name' and 'season' (keep from 3rd game onward)
df_new = (
    df.groupby(['name', 'season'])
    .apply(lambda x: x.iloc[2:])
    .reset_index(drop=True)
)

# Keep only key columns and moving averages
moving_avg_cols = [col for col in df_new.columns if '_3MA' in col]
object_cols = ['name', 'season', 'Game_Number']

# Ensure no duplicate columns
final_cols = list(dict.fromkeys(object_cols + moving_avg_cols))
df_moving_avg = df_new[final_cols]

# Check the result
print(df_moving_avg.head())

```

```python
import pandas as pd
import numpy as np

def high_correlation_vars(df, threshold=0.9):
    corr_matrix = df_numeric_combined.corr()
    mask = np.tril(np.ones(corr_matrix.shape), k=0).astype(bool)
    filtered_corr = corr_matrix.mask(mask)

    pairs = (
        filtered_corr.stack()
        .reset_index()
        .rename(columns={'level_0': 'Variable_1', 'level_1': 'Variable_2', 0: 'Correlation'})
        .sort_values(by='Correlation', ascending=False)
        .reset_index(drop=True)
    )

    return pairs[pairs['Correlation'] > threshold]

top_correlations = high_correlation_vars(df)
print(top_correlations)
```
