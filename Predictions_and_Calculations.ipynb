```python
#Random Forest

bet_amount = 100

df_RF["Bet_Placed"] = df_RF["Propensity_RF"] > 0.7

df_RF["Bet_Outcomes"] = 0

for idx, row in df_RF.iterrows():
    if row["Bet_Placed"]:  # If bet is placed
        if row["win_indicator"] == row["Prediction_RF"]:
            if row["OT"] == "Yes":  # If OT is "Yes", it's a loss
                df_RF.at[idx, "Bet_Outcomes"] = -bet_amount
            else:  # If correct prediction and not OT, calculate net winnings
                df_RF.at[idx, "Bet_Outcomes"] = (row["Decimal_Odds"] - 1) * bet_amount
        else:  # If incorrect prediction, it's a loss
            df_RF.at[idx, "Bet_Outcomes"] = -bet_amount

# Calculate total amount bet
total_bet_amount = df_RF["Bet_Placed"].sum() * bet_amount

# Calculate final profit/loss
total_profit_loss = df_RF["Bet_Outcomes"].sum()

# Calculate ROI
roi = (total_profit_loss / total_bet_amount) if total_bet_amount > 0 else 0

# Print results
print(f"Total Amount Bet: ${total_bet_amount:.2f}")
print(f"Total Profit/Loss: ${total_profit_loss:.2f}")
print(f"ROI: {roi:.2%}")

```

```python
import pandas as pd


bet_amount = 100

cutoffs = [0.5, 0.6, 0.7, 0.8]


results = []

for cutoff in cutoffs:
    # Create a boolean column for whether a bet is placed
    df_XGB["Bet_Placed"] = df_XGB["Propensity_XGB"] > cutoff

    # Initialize outcome column
    df_XGB["Bet_Outcomes"] = 0

    # Calculate bet outcomes
    for idx, row in df_XGB.iterrows():
        if row["Bet_Placed"]:  # If bet is placed
            if row["win_indicator"] == row["Prediction_XGB"]:
                if row["OT"] == "Yes":  # If OT is "Yes", it's a loss
                    df_XGB.at[idx, "Bet_Outcomes"] = -bet_amount
                else:  # If correct prediction and not OT, calculate net winnings
                    df_XGB.at[idx, "Bet_Outcomes"] = (row["Decimal_Odds"] - 1) * bet_amount
            else:  # If incorrect prediction, it's a loss
                df_XGB.at[idx, "Bet_Outcomes"] = -bet_amount

    # Calculate key metrics
    num_games_bet = df_XGB["Bet_Placed"].sum()
    total_bet_amount = num_games_bet * bet_amount
    total_profit_loss = df_XGB["Bet_Outcomes"].sum()
    roi = (total_profit_loss / total_bet_amount) if total_bet_amount > 0 else 0

    # Store results
    results.append([cutoff, num_games_bet, total_bet_amount, total_profit_loss, roi * 100])

# Create a results dataframe
df_results = pd.DataFrame(results, columns=["Propensity Cutoff", "Games Bet", "Total Wagered", "Total Revenue", "ROI (%)"])

# Print results table
print(df_results)
```
